```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
traffic = [150, 200, 180, 220, 250, 210]
```

```python
plt.plot(months,traffic)
plt.xlabel("Months")
plt.ylabel("Monthly traffic in K")
plt.title("Traffic Data")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(months,traffic,marker='o',linestyle='--',color='b')
plt.xlabel("Months")
plt.ylabel("Monthly traffic in K")
plt.title("Traffic Data")
plt.grid(True)
plt.show()
```

```python
### Plot Types
### Bar chart
# Expense categories
categories = ['Housing', 'Transportation', 'Food', 'Entertainment', 'Utilities']

# Monthly expenses for Alice, Bob, and Carol
alice_expenses = [1200, 300, 400, 200, 150]
bob_expenses = [1100, 320, 380, 180, 140]
carol_expenses = [1300, 280, 420, 220, 160]
bar_width = 0.2
x = np.arange(len(categories))
print(x)
plt.figure(figsize=(8,6))
print(x - bar_width)
plt.bar(x - bar_width, alice_expenses, width=bar_width,color='b', label='Alice')
plt.bar(x,bob_expenses, width=bar_width, color='g', label='Bob')
plt.bar(x + bar_width, carol_expenses, width=bar_width,color='y', label='Carol')
plt.xlabel('Category')
plt.ylabel('Expenses')
plt.title('Individual Expenses by Category')
plt.xticks(x,categories)
plt.legend()
plt.show()
```

```python
### Histograms
# Sample exam scores data
exam_scores = [68, 72, 75, 80, 82, 84, 86, 90, 92, 95, 98, 100]
# Custom bin ranges
bin_ranges = [60, 70, 80, 90, 100]
plt.figure(figsize=(6,4))
plt.hist(exam_scores,bins=bin_ranges,color='green', edgecolor='black',alpha=0.5)
median = np.median(exam_scores)
plt.axvline(median,color='red',linestyle='--',linewidth=2,label=f'Median Score: {median}')
plt.xlabel("Exam Scores")
plt.ylabel("Freq")
plt.title("Scores Histogram")
plt.legend()
plt.show()
```

```python
### Scatter plots
# Let's take a sample data for stores
# Store Size Represents the size of each store in 100sq.ft
stores = ['Store A', 'Store B', 'Store C', 'Store D', 'Store E']
customers = [120, 90, 150, 80, 200]
revenue = [20000, 18000, 25000, 17000, 30000]
store_size = [10, 5, 15, 8, 20] 
points = [size*100 for size in store_size]
plt.figure(figsize=(6,4))
plt.scatter(customers,revenue,s=points,c='skyblue', alpha=0.7, edgecolors='b')
plt.xlabel("# Cust")
plt.ylabel("Revenue $USD")
plt.title("Relationship b/w Cust, Revenue, Store Size")
plt.show()
```

```python
### Pie Chart
# Product categories
categories = ['Electronics', 'Clothing', 'Home Decor', 'Books', 'Toys']

# Sales data for each category
sales = [3500, 2800, 2000, 1500, 1200]

plt.figure(figsize=(5,5))
plt.pie(sales,labels=categories,explode=(0,0.1,0,0,0),shadow=True,autopct=f'%1.1f%%')
plt.title("Sales Chart")
plt.show()
```

```python
### Box Plot
np.random.seed(42)
data1 = np.concatenate([np.random.normal(0, 1, 100), np.random.normal(6, 1, 10)])
plt.figure(figsize=(5,5))
plt.boxplot(data1,vert=False,patch_artist=True,boxprops={'facecolor': 'lightblue'},flierprops={'marker': 'o', 'markerfacecolor': 'red', 
                        'markeredgecolor': 'red'})
plt.title("Box Plot")
plt.show()
```

```python
### Heat Map
correlation_matrix = np.array([[1.0, 0.8, 0.3, -0.2],
                               [0.8, 1.0, 0.5, 0.1],
                               [0.3, 0.5, 1.0, -0.4],
                               [-0.2, 0.1, -0.4, 1.0]])
plt.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1, aspect='auto', origin='upper')
cbar = plt.colorbar()
cbar.set_label('Correlation', rotation=270, labelpad=20)
plt.title('Correlation Matrix Heatmap')
plt.xticks(range(len(correlation_matrix)), ['Var1', 'Var2', 'Var3', 'Var4'])
plt.yticks(range(len(correlation_matrix)), ['Var1', 'Var2', 'Var3', 'Var4'])

plt.show()
```

```python
# Sample data for stack plot
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
electronics = [10000, 12000, 11000, 10500]
clothing = [5000, 6000, 7500, 8000]
home_appliances = [7000, 7500, 8200, 9000]

# Create a stack plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.stackplot(quarters, electronics, clothing, home_appliances, labels=['Electronics', 'Clothing', 'Home Appliances'],
              colors=['blue', 'green', 'red'], alpha=0.7)

# Add labels, legend, and title
plt.xlabel('Quarters')
plt.ylabel('Sales ($)')
plt.title('Product Category Sales Over Quarters')
plt.legend(loc='upper left')

# Display the plot
plt.grid(True)
plt.show()
```
